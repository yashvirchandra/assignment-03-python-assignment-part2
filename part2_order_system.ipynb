```python
# =========================================
#  Provided/ Given data for the restaurant management system
# ========================================
 
menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":       {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}

inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  5, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}

sales_log = {
    "2025-01-01": [
        {"order_id": 1,  "items": ["Paneer Tikka", "Garlic Naan"],          "total": 220.0},
        {"order_id": 2,  "items": ["Gulab Jamun", "Veg Soup"],              "total": 210.0},
        {"order_id": 3,  "items": ["Butter Chicken", "Garlic Naan"],        "total": 360.0},
    ],
    "2025-01-02": [
        {"order_id": 4,  "items": ["Dal Tadka", "Garlic Naan"],             "total": 220.0},
        {"order_id": 5,  "items": ["Veg Biryani", "Gulab Jamun"],           "total": 340.0},
    ],
    "2025-01-03": [
        {"order_id": 6,  "items": ["Paneer Tikka", "Rasgulla"],             "total": 260.0},
        {"order_id": 7,  "items": ["Butter Chicken", "Veg Biryani"],        "total": 570.0},
        {"order_id": 8,  "items": ["Garlic Naan", "Gulab Jamun"],           "total": 130.0},
    ],
    "2025-01-04": [
        {"order_id": 9,  "items": ["Dal Tadka", "Garlic Naan", "Rasgulla"], "total": 300.0},
        {"order_id": 10, "items": ["Paneer Tikka", "Gulab Jamun"],          "total": 270.0},
    ],
}
```

```python
# =========================================
#  TASK 1 
# =========================================

# Finding all categories First
categories = ["Starters", "Mains", "Desserts"]

# Now Printing Goruped menu
for category in categories:
    print("====="+ category + "=====")
    for item_name, details in menu.items():
        if details["category"] == category:
            status = "Available" if details["available"] else "Not Available"
            print(f"{item_name:15} ₹{details['price']:6.2f} - [{status}] ")
    print()

# Total items
print(f"Total number of items in the menu: {len(menu)}")

# Available items= [
available_count = sum(1 for item in menu.values() if item["available"])
print(f"Available items: {available_count}")

# Most Expensive Item In Menu
most_expensive = max(menu.items(), key=lambda x: x[1]["price"])
print("Most expensive item: ", most_expensive[0], " at ₹", most_expensive[1]["price"], "price: ")

# Items Under ₹150
print("Items under ₹150:")
for item_name, details in menu.items():
    if details["price"] < 150:
        print(f"{item_name:15} ₹{details['price']:6.2f}")


```

```python
# =========================================
# TASK 2
# =========================================
cart = []

#Adding items to cart
def add_to_cart(item_name,quantity):
    if item_name not in menu:
        print(item_name, "Item does not exist in menu.")
        return
    
    # Checking if item is available
    if not menu[item_name]["available"]:
        print(item_name, "Item unavailable.")
        return
    
    # Check If item is already in Cart,
    for item in cart:
        if item["item"] == item_name:
            item["quantity"] += quantity
            return
        
    # Adding new item to Cart
    cart.append({
        "item": item_name,
        "quantity": quantity,
        "price": menu[item_name]["price"]
    })

# Removing item from cart
def remove_from_cart(item_name):
    for item in cart:
        if item["item"] == item_name:
            cart.remove(item)
            return
    print( item_name ,"Removed from cart.")

# Simulate Steps
add_to_cart("Paneer Tikka", 2)
add_to_cart("Gulab jamun", 1)
add_to_cart("Paneer Tikka", 1) # Update quantity
add_to_cart("Mystery Burger", 1) # Non Existent Item
add_to_cart("Chicken Wings", 1) # Unavailable Item
remove_from_cart("Gulab jamun") # Remove item from cart

print("\nCart after operations: ", cart)

# Order Summary
subtotal= sum(item["quantity"] * item["price"] for item in cart)
gst = subtotal * 0.05
total = subtotal + gst

print("\n========== Order Summary ==========")

for item in cart:
    total_price = item["quantity"] * item["price"]

    print(f"{item['item']:20} x {item['quantity']:<3} ₹{total_price:7.2f}")

print("-----------------------------------")

print(f"{'Subtotal:':25} ₹{subtotal:7.2f}")
print(f"{'GST (5%):':25} ₹{gst:7.2f}")
print(f"{'Total Payable:':25} ₹{total:7.2f}")

print("===================================")
```

```python
# =================================
# TASK 3
# =================================
import copy

#Creating A deep Copy of Inventory
inventory_backup = copy.deepcopy(inventory)

# Working of deep copy displayed
print("\n----- Deep Copy Verification -----")
inventory["Paneer Tikka"]["stock"] = 10 #temporarily changing stock to show deep copy effect

print("\nAfter chaning original inventory: ")
print("Inventory: ", inventory["Paneer Tikka"])
print("Backup: ", inventory_backup["Paneer Tikka"])

# Restoring original inventory for Task 3
inventory = copy.deepcopy(inventory_backup)

print("\nInventory Restored.")

# Deduct Stock from inventory of Task 2
print("\nUpdating Inventory after Order...\n")

for item in cart:
    name = item["item"]
    quantity = item["quantity"]

    if inventory[name]["stock"] < quantity:
        print("Low Stock for ", name)
        inventory[name]["stock"] = 0
    else:
        inventory[name]["stock"] -= quantity

# Printing Updated Inventory
print("\n----- Updated Inventory -----")

for item, details in inventory.items():
    print(f"{item}: Stock = {details['stock']} units left")

# Reorder Alert
print("\nReorder Alerts:")

for item, details in inventory.items():
    if details["stock"] <= details["reorder_level"]:
        print(f"⚠ Reorder Alert: {item} — Only {details['stock']} left (reorder level: {details['reorder_level']})")

# Final check
print("\n----- Final Inventory Check (Deep Copy Proof) -----")
print("Current Inventory Paneer Tikka: ", inventory["Paneer Tikka"])
print("Backup Inventory Paneer Tikka: ", inventory_backup["Paneer Tikka"])
```

```python
# ==============================
# TASK 4
# ==============================
print("\n----- Sales Log Analysis -----")

# Calculating Total Revenue per day
revenue_per_day = {} # Dictionary to store revenue per day

for date, orders in sales_log.items():
    
    #Sum of all orders for the day
    total_revenue = sum(order["total"] for order in orders)
    revenue_per_day[date] = total_revenue

# Printing Revenue per day
for date, total in revenue_per_day.items():
    print(f"{date}: ₹{total:.2f}") 

# Finding Best Selling Item of the Day
best_day = max(revenue_per_day, key=revenue_per_day.get)
print(f"\nBest Selling Day: ")
print(f"{best_day} with revenue of ₹{revenue_per_day[best_day]:.2f}")

# Finding Most Ordered Item
item_count = {} # Dictionary to count item orders   

for orders in sales_log.values():
    for order in orders:
        for item in order["items"]:
            
            # Counting each item appearance
            item_count[item] = item_count.get(item, 0) + 1

# Finding Most Ordered Item
most_ordered_item = max(item_count, key=item_count.get)

print("\nMost ordered item: ")
print(f"{most_ordered_item} (ordered {item_count[most_ordered_item]} times)")

# Adding new Day Data
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]
            
print("\nNew day Added : 2025-01-05")

# Recalculating Revenue after Update
updated_revenue = {}

for date, orders in sales_log.items():
    updated_revenue[date] = sum(order["total"] for order in orders)

print("\nUpdated revenue per day: ")
for date, total in updated_revenue.items():
    print(f"{date}: ₹{total:.2f}")

# Finding New Best Selling Day
updated_best_day = max(updated_revenue, key=updated_revenue.get)

print("\nUpdated Best Selling Day: ")
print(f"{updated_best_day} with ₹{updated_revenue[updated_best_day]:.2f}")

# Printing All Orders Using Enumerate 
print("\nAll Orders:\n")

counter = 1 # ManualNumbering

for date, orders in sales_log.items():
    for order in orders: 
        items_list = ", ".join(order["items"]) # Converting list of items to string

        print(f"{counter}. [{date}] Order #{order['order_id']} - ₹{order['total']:.2f} - Items: {items_list}")

        counter += 1
```
